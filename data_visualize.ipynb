Subplots (2 row, 2 columns)

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [5, 15, 7, 10]
data = np.random.randn(1000)

# ::- PLOT 1: Top Left ::
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b-')
plt.title("Line Plot")

# ::- PLOT 2: Top Right ::
plt.subplot(2, 2, 2)
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")

# ::- PLOT 3: Bottom Left ::
plt.subplot(2, 2, 3)
plt.bar(categories, values, color='green')
plt.title("Bar Chart")

# ::- PLOT 4: Bottom Right ::
plt.subplot(2, 2, 4)
plt.hist(data, bins=20, color='orange')
plt.title("Histogram")

plt.tight_layout()
plt.savefig("plot.png",dpi = 300,bbox_inches = "tight")
```

Problem 1: Plot y = x²

```python
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x,y,"r--",label = "y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Function")
plt.legend()
```

Problem 2: 2×2 Subplot Grid

```python
x = np.linspace(0.1,10,200)  # avoid x=0 for 1/x
plt.figure(figsize=(8,6))
#y =x
plt.subplot(2,2,1)
plt.plot(x,x,"b")
plt.title("y = x")
#y = x^2
plt.subplot(2,2,2)
plt.plot(x,x**2,"r")
plt.title("y = x^2")
# y = sqrt(x)
plt.subplot(2,2,3)
plt.plot(x,np.sqrt(x),"g")
plt.title("y=sqrt(x)")
#y = 1 / x
plt.subplot(2,2,4)
plt.plot(x,1/x,"m")
plt.title("y = 1/x")

plt.tight_layout()
```

Problem 3: Bar Chart with Labels

```python
sales = [50, 70, 30, 90, 60]
categories = ['A','B','C','D','E']

plt.bar(categories, sales, color='skyblue')

# Add value labels on bars
for i, v in enumerate(sales):
    plt.text(i, v+2, str(v), ha='center')

plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales Comparison")
plt.show()
```

Problem 4: Histogram with Mean Line

```python
data = np.random.normal(0, 1, 500)  # mean=0, std=1

plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.axvline(data.mean(), color='blue', linestyle='--', label='Mean')

plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution Histogram")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.randn(100)
y = np.random.randn(100)

plt.scatter(x,y,c=y,s=x*500,alpha=0.7)
plt.colorbar(label="Y Value")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Custom scatter plot")
```

```python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
plt.plot(x,y)
max_x = np.pi / 2
max_y = 1
plt.scatter(max_x,max_y)
plt.annotate("Maximum",xy = (max_x,max_y),xytext=(2,0.8),arrowprops=dict())
plt.grid(True)
```

```python
data = np.random.normal(50,10,100)
mean=  np.mean(data)
median = np.median(data)
plt.hist(data,bins=30)
plt.axvline(mean,color="red",linestyle="--",label="mean")
plt.axvline(median,color ="black", linestyle=":",label="Median")
plt.legend()
```
