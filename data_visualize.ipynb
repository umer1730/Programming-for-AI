```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0,4*np.pi,200)
y_sin = np.sin(x)
plt.plot(x, y_sin, "b-",label = "sin")
```

```python
plt.scatter(x[::10],y_sin[::10],color = "red",label="cos")
plt.legend()
```

```python
plt.plot(x, y_sin, "b-",label = "sin")
plt.scatter(x[::10],y_sin[::10],color = "red",label="cos")
plt.legend()
```
