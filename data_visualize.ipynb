```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
categories = ['A', 'B', 'C', 'D']
values = [5, 15, 7, 10]
data = np.random.randn(1000)

# ::- PLOT 1: Top Left ::
plt.subplot(2, 2, 1)
plt.plot(x, y, 'b-')
plt.title("Line Plot")

# ::- PLOT 2: Top Right ::
plt.subplot(2, 2, 2)
plt.scatter(x, y, color='red')
plt.title("Scatter Plot")

# ::- PLOT 3: Bottom Left ::
plt.subplot(2, 2, 3)
plt.bar(categories, values, color='green')
plt.title("Bar Chart")

# ::- PLOT 4: Bottom Right ::
plt.subplot(2, 2, 4)
plt.hist(data, bins=20, color='orange')
plt.title("Histogram")

plt.tight_layout()
plt.savefig("plot.png",dpi = 300,bbox_inches = "tight")
```

```python
x = np.linspace(-10, 10, 100)
y = x**2
plt.plot(x,y,"r--",label = "y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Quadratic Function")
plt.legend()
```

```python
x = np.linspace(0.1,10,200)
plt.figure(figsize=(8,6))

plt.subplot(2,2,1)
plt.plot(x,x,"b")
plt.title("y = x")

plt.subplot(2,2,2)
plt.plot(x,x**2,"r")
plt.title("y = x^2")

plt.subplot(2,2,3)
plt.plot(x,np.sqrt(x),"g")
plt.title("y=sqrt(x)")

plt.subplot(2,2,4)
plt.plot(x,1/x,"m")
plt.title("y = 1/x")

plt.tight_layout()
```

```python
sales = [50, 70, 30, 90, 60]
categories = ['A','B','C','D','E']

plt.bar(categories, sales, color='skyblue')

# Add value labels on bars
for i, v in enumerate(sales):
    plt.text(i, v+2, str(v), ha='center')

plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales Comparison")
plt.show()
```
