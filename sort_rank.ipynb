Sorting & Ranking

```python
import pandas as pd

data = {
    "Name": ["Ali","Sara","Ahmed","Zara","Usman","Hina","Bilal","Ayesha"],
    "Age": [21,23,20,22,24,21,23,22],
    "City": ["Lahore","Karachi","Lahore","Islamabad","Karachi","Lahore","Islamabad","Karachi"],
    "Marks": [85,90,78,92,88,76,95,89]
}

df = pd.DataFrame(data)
print(df)
```

df.sort_values() → Sort by column
Sort by Marks (ascending)

```python
print(df.sort_values("Marks")) 
```

```python
# sort values by name index
print(df.sort_values("Name")) 
```

```python
# sort by marks descending
print(df.sort_values("Marks",ascending=False))
```

```python
# sort by multiple column
print(df.sort_values(["Age","Marks"]))
```

df.sort_index() → Sort by row index

```python
# reverse index order
print(df.sort_index(ascending=False))
```

df.rank() → Ranking values

Creates ranking numbers

```python
# rank marks
df["Rank"] = df["Marks"].rank()
df
```

```python
# highest rank marks equal to 1
df["Rank"] = df["Marks"].rank(ascending=False)
df
```

df.nlargest() → Top values
Top 3 students by Marks

```python
print(df.nlargest(3,"Marks"))
```

df.nsmallest() → Lowest values
Lowest 3 Marks

```python
print(df.nsmallest(3,"Marks"))
```

```python
# A function to categorize sensor readings
def evaluate_voltage(v):
    if v > 14.0:
        return "High"
    elif v < 11.0:
        return "Low"
    else:
        return "Normal"
```

```python
import pandas as pd

df = pd.DataFrame({
    "Voltage": [10.5, 12.0, 15.2, 9.8, 13.5]
})

df["Status"] = df["Voltage"].apply(evaluate_voltage)
print(df)
```

```python
df[["Voltage","Status"]].head()
```

GroupBy & Apply

```python
import pandas as pd

data = {
    "Name": ["Ali","Sara","Ahmed","Zara","Usman","Hina","Bilal","Ayesha"],
    "City": ["Lahore","Karachi","Lahore","Islamabad","Karachi","Lahore","Islamabad","Karachi"],
    "Department": ["IT","HR","IT","Finance","HR","IT","Finance","HR"],
    "Marks": [85,90,78,92,88,76,95,89],
    "Salary": [50000,60000,45000,65000,62000,48000,70000,61000]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.groupby("City")["Marks"].mean()
```

```python
df.groupby("City")["Marks"].agg(["mean","max","min"])
```

```python
df.groupby(["City","Department"])["Salary"].mean()
```

```python
df.groupby("Department").mean(numeric_only=True)
```

```python
df.groupby("City").size()
```

Or

```python
df.groupby("City")["Name"].count()
```

```python
groups = df.groupby("City")

for city, data in groups:
    print("\nCity:",city)
    print(data)
```

```python
def bonus(x):
    return x + 5000
df.groupby("Department")["Salary"].apply(bonus)
```

```python
df["Salary"]
```

```python
df["City_Avg_Marks"] = df.groupby("City")["Marks"].transform("mean")
df
```

```python
#Keep cities where avg marks > 85
df.groupby("City").filter(lambda x:  x["Marks"].mean() > 85)
```
