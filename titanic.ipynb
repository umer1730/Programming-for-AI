Boolean Indexing

```python
mask = df["Pclass"] == 1 # is place equal to 1
first_class = df[mask] # new DataFrame with only 1st class passengers
first_class
```

```python
winners = df[(df["Sex"] == "female") & (df["Survived"] == 1)]
winners
```

```python
not_third = df[~(df["Pclass"] == 3)]
not_third
```

```python
subset = df[(df["Pclass"] == 2) | (df["Pclass"] == 3)]
subset

```

```python
find_male = df[(df["Sex"] == "male") & (df["Survived"] == 1)]
find_male
```

```python
target_ports = ["C","Q"]
subset = df[df["Embarked"].isin(target_ports)]
subset
```

```python
subset = df[df["Pclass"].isin([1,2])]
subset
```

```python
name_search = df[df["Name"].str.contains("William",na=False)]
name_search
```
